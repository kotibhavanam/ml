```python
##replacing the value of 88 with 8
df=df.replace(88,8)
```

```python
# class = lambda df['score']: 0 if df['score'] <= 5 else 1 (using lamda function to create two classes based on the condition)
df["class"]=df['score'].apply(lambda x: 0 if x <= 5 else 1)
df
```

```python
import numpy as np
import numpy as np
# Assuming you have 'n' classes and each class is stored in a list
class1_vectors = df[df['class']== 0]
class2_vectors = df[df['class']== 1]
```

```python
#checking shape of the class1
class1_vectors.shape
```

```python
#droping the coloums score and class in class1 vector
class1_vectors.drop(['score', 'class'], axis=1, inplace=True)
```

```python
#checking the shape of dataset after droping 
class1_vectors.shape
```

```python
#droping the coloums score and class in class2 vector
class2_vectors.drop(['score', 'class'], axis=1, inplace=True)
```

```python
#checking the shape of dataset after droping 
class2_vectors.shape
```

```python
# Calculate the mean (class centroid) for each class
class1_mean = np.mean(class1_vectors, axis=0)
class2_mean = np.mean(class2_vectors, axis=0)
#printing the mean of class1 and class2
print("Mean for Class 1:", class1_mean)
print("Mean for Class 2:", class2_mean)
```

```python
#calucating the stantard deviation of class1 and class2
std_class1 = np.std(class1_vectors, axis=0)
std_class2 = np.std(class2_vectors, axis=0)
#printing the calucated standard diviation
print("Standard Deviation for Class 1:", std_class1)
print("Standard Deviation for Class 2:", std_class2)
```

```python
#finding the distance between two classes using centroid
distance = np.linalg.norm(class1_mean - class2_mean)
#printing the calucated distance
print("Distance between centroids:", distance)
```

```python
#dividing dataset into two datas 
X = df.iloc[:, 0:767]
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
#diving the datas into two (train and test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(X_train, y_train)
```

```python
#to caluculate the accuracy
accuracy = neigh.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
#predicting the lables
predicted_labels = neigh.predict(X_test)
print (predicted_labels )
```

```python
nn_accuracies = []
knn_accuracies = []

k_values = list(range(1, 12))

for k in k_values:
    # Nearest Neighbor Classifier (k = 1)
    nn_classifier = KNeighborsClassifier(n_neighbors=1)
    nn_classifier.fit(X_train, y_train)
    nn_accuracy = nn_classifier.score(X_test, y_test)
    nn_accuracies.append(nn_accuracy)
    
    
    knn_classifier = KNeighborsClassifier(n_neighbors=3)
    knn_classifier.fit(X_train, y_train)
    knn_accuracy = knn_classifier.score(X_test, y_test)
    knn_accuracies.append(knn_accuracy)
    
plt.figure(figsize=(10, 6))
plt.plot(k_values, nn_accuracies, marker='o', label='NN (k=1) Accuracy')
plt.plot(k_values, knn_accuracies, marker='o', label='k-NN (k=3) Accuracy')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k Value for NN and k-NN')
plt.legend()
plt.grid(True)
plt.xticks(k_values)
plt.show()    
```

```python
from sklearn.metrics import confusion_matrix

# Predict on the training and test data
y_train_pred_nn = nn_classifier.predict(X_train)
y_test_pred_nn = nn_classifier.predict(X_test)

y_train_pred_knn = knn_classifier.predict(X_train)
y_test_pred_knn = knn_classifier.predict(X_test)

# Compute the confusion matrix for training and test data
confusion_matrix_train_nn = confusion_matrix(y_train, y_train_pred_nn)
confusion_matrix_test_nn = confusion_matrix(y_test, y_test_pred_nn)

confusion_matrix_train_knn = confusion_matrix(y_train, y_train_pred_knn)
confusion_matrix_test_knn = confusion_matrix(y_test, y_test_pred_knn)

print("Confusion Matrix for NN (Training Data):")
print(confusion_matrix_train_nn)

print("Confusion Matrix for NN (Test Data):")
print(confusion_matrix_test_nn)

print("Confusion Matrix for k-NN (Training Data):")
print(confusion_matrix_train_knn)

print("Confusion Matrix for k-NN (Test Data):")
print(confusion_matrix_test_knn)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision, recall, and F1-score for NN
precision_train_nn = precision_score(y_train, y_train_pred_nn, average='weighted')
recall_train_nn = recall_score(y_train, y_train_pred_nn, average='weighted')
f1_score_train_nn = f1_score(y_train, y_train_pred_nn, average='weighted')

precision_test_nn = precision_score(y_test, y_test_pred_nn, average='weighted')
recall_test_nn = recall_score(y_test, y_test_pred_nn, average='weighted')
f1_score_test_nn = f1_score(y_test, y_test_pred_nn, average='weighted')

# Calculate precision, recall, and F1-score for k-NN
precision_train_knn = precision_score(y_train, y_train_pred_knn, average='weighted')
recall_train_knn = recall_score(y_train, y_train_pred_knn, average='weighted')
f1_score_train_knn = f1_score(y_train, y_train_pred_knn, average='weighted')

precision_test_knn = precision_score(y_test, y_test_pred_knn, average='weighted')
recall_test_knn = recall_score(y_test, y_test_pred_knn, average='weighted')
f1_score_test_knn = f1_score(y_test, y_test_pred_knn, average='weighted')

print("Precision for NN (Training Data):", precision_train_nn)
print("Recall for NN (Training Data):", recall_train_nn)
print("F1-Score for NN (Training Data):", f1_score_train_nn)

print("Precision for NN (Test Data):", precision_test_nn)
print("Recall for NN (Test Data):", recall_test_nn)
print("F1-Score for NN (Test Data):", f1_score_test_nn)

print("Precision for k-NN (Training Data):", precision_train_knn)
print("Recall for k-NN (Training Data):", recall_train_knn)
print("F1-Score for k-NN (Training Data):", f1_score_train_knn)

print("Precision for k-NN (Test Data):", precision_test_knn)
print("Recall for k-NN (Test Data):", recall_test_knn)
print("F1-Score for k-NN (Test Data):", f1_score_test_knn)
```
