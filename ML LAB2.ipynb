```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.metrics import pairwise_distances

 

# Load your data from the Excel file

file_name = "lab2.xlsx"

worksheet_name = "marketing_campaign"

data = pd.read_excel(file_name, sheet_name=worksheet_name)

 

# Select the first 20 observation vectors

selected_data = data.iloc[:20, 1:-1]  # Exclude the first column (ID) and the last column (Response)

 

# Replace missing values '?' with NaN

selected_data = selected_data.replace('?', np.nan)

 

# Fill missing values with 0 (you can choose a different strategy if needed)

selected_data = selected_data.fillna(0)

 

# Identify columns with non-numeric values (e.g., 'Education' and 'Marital_Status')

non_numeric_columns = selected_data.select_dtypes(exclude=[np.number]).columns

 

# Convert non-numeric columns to numeric using one-hot encoding

selected_data = pd.get_dummies(selected_data, columns=non_numeric_columns)

 

# Calculate Jaccard Coefficient (JC) between pairs of vectors

jc_matrix = 1 - pairwise_distances(selected_data, metric='hamming')

 

# Calculate Simple Matching Coefficient (SMC) between pairs of vectors

smc_matrix = 1 - pairwise_distances(selected_data, metric='matching')

 

# Create a heatmap to visualize the similarities

sns.set(style="white")

plt.figure(figsize=(10, 4))

 

# Jaccard Coefficient Heatmap

plt.subplot(121)

sns.heatmap(jc_matrix, annot=True, cmap="YlGnBu", cbar=False, xticklabels=False, yticklabels=False)

plt.title("Jaccard Coefficient")

 

# Simple Matching Coefficient Heatmap

plt.subplot(122)

sns.heatmap(smc_matrix, annot=True, cmap="YlGnBu", cbar=False, xticklabels=False, yticklabels=False)

plt.title("Simple Matching Coefficient")

 

plt.tight_layout()

plt.show()
```
