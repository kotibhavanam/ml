```python
import numpy as np
import matplotlib.pyplot as plt

# Step activation function
def step(x):
    return 1 if x >= 0 else 0

# Training data for AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
W = np.array([10, 0.2, -0.75])
learning_rate = 0.05

# Lists to store error and epochs
error_history = []
epochs = 0

# Training loop
while True:
    error = 0
    for i in range(len(X)):
        # Calculate the weighted sum
        weighted_sum = np.dot(X[i], W)
        
        # Apply the Step activation function
        output = step(weighted_sum)
        
        # Calculate the error
        delta = y[i] - output
        
        # Update weights and bias
        W += learning_rate * delta * X[i]
        
        # Calculate and accumulate the error (sum-square-error)
        error += delta ** 2
    
    # Append the error to the history
    error_history.append(error)
    
    # Check for convergence (error is zero for all training samples)
    if error == 0:
        break
    
    epochs += 1

# Plot the epochs against the error values
plt.plot(range(epochs + 1), error_history)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error vs. Epochs')
plt.show()

print(f"Converged in {epochs} epochs")
print("Final Weights:")
print(f"W0: {W[0]}, W1: {W[1]}, W2: {W[2]}")
```
